```python
with open('cards.csv', 'r') as cards:
    cards = cards.read()
```

```python
def clean_data(inp):
    lines = inp.split('\n')[:-1]
    return lines
```

```python
import re
def check_type(inp):
    card_dict = {}
    for char in inp:
        if not card_dict.get(char):
            card_dict[char] = 1
        else:
            card_dict[char] += 1
    print(card_dict)
# Five of a kind, where all five cards have the same label: AAAAA
    if max(card_dict.values()) == 5:
        return 7
# Four of a kind, where four cards have the same label and one card has a different label: AA8AA
    elif max(card_dict.values()) == 4:
        return 6
# Full house, where three cards have the same label, and the remaining two cards share a different label: 23332
    elif len(card_dict.keys()) == 2:
        return 5
# Three of a kind, where three cards have the same label, and the remaining two cards are each different from any other card in the hand: TTT98
    elif max(card_dict.values()) == 3:
        return 4
# Two pair, where two cards share one label, two other cards share a second label, and the remaining card has a third label: 23432
    elif len(card_dict.keys()) == 3:
        return 3
# One pair, where two cards share one label, and the other three cards have a different label from the pair and each other: A23A4
    elif len(card_dict.keys()) == 4:
        return 2
# High card, where all cards' labels are distinct: 23456
    else:
        return 1
```

```python
card_lines = clean_data(cards)
```

```python
# Get card type to rank on the macro level first
card_type_dict = {i:[] for i in range(1,8)}
for card_line in card_lines:
    card, bet = card_line.split()[0], card_line.split()[1]
    hand_type = check_type(card)
    card_type_dict[hand_type].append(card_line)
```

```python
def rank_same_type(inp):
    suite_dict = {'2': 1, '3': 2, '4': 3, '5': 4, '6':5, '7':6, '8':7, '9':8, 'T':9, 'J':10, 'Q': 11, 'K':12, 'A':13}
    return [suite_dict[char] for char in inp]
```

```python
# now start from lowest card type 1:
total_count = 0
rank_counter = 1
for card_type, cards_details in card_type_dict.items():
    # card_type is already in order, starts from 1
    # start checking the cards and get the lowest first value.
    ranking_dict = {}
    for card_details in cards_details:
        card, bet = card_details.split()[0], card_details.split()[1]
        ranking_dict[bet] = rank_same_type(card)
    ranking_list = sorted(ranking_dict.items(), key=lambda x: x[1])
    for card_tuple in ranking_list:
        total_count += rank_counter*int(card_tuple[0])
        rank_counter += 1
print(total_count)
```
